```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
# Data
data = [ ("Java", "20000"), ("Python", "100000"), ("Scala", "3000" )]
# Columns
columns = [ "language", "users_count" ]
# Create DataFrame
df = spark.createDataFrame(data). toDF (*columns)
```

```python
df.show()
```
